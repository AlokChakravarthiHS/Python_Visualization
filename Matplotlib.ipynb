# MatplotLib 

it is a Visualization technique we use in python ,
We use it for creating graphs 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)

```

## Plotting 

```python
#using scatter 
plt.scatter(x,y,c='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter graph')
plt.show()
```

```python
#we can save the any particular file by using
plt.savefig('pattern_name.jpg/png') 
```

```python
#plot
y=x*x
plt.plot(x,y,c='r',marker='*',linestyle="dashed",linewidth=1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('y=x*x')
plt.show()
```

```python
#creating Subplots
x=np.arange(0,4*np.pi,0.1)
y=np.sinc(x)
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
plt.subplot(2,2,1)
plt.plot(x,y_sin,c="g")
plt.title('sine wave')
plt.show()

x2=np.arange(0,4*np.pi,0.1)
y_cos=np.cos(x2)
plt.subplot(2,2,2)
plt.plot(x2,y_cos,c='r')
plt.title('cosine wave')
plt.show()
```

```python
#BarPlot

x=[2,8,18]
y=[5,7,23]
x2=[6,10,43]
y2=[7,9,12]

plt.bar(x,y)
plt.bar(x2,y2)
plt.show()
```

```python
# Histogram
a=[22,34,56,46,38,55,56,54,23,98,86,68]
plt.hist(a)
plt.show()

#the plot here is done with respect to the bins bins is grouping data into equal width
#x-axis takes value of input, y-axis gives the denity or the number of elements in the rannge og values "a"
```

```python
# BoxPlot
#It hels us to find out percentiles
data=[np.random.normal(0,std,100)for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
plt.title('boxplot')
#oo in the graph is outliers this comes when we randomly pick up values 
#percentiles are in range of 0,25,50,75,100 for all values
```

```python
# Pie Chart\

label=['java','python','C++','react']
size=[213,534,654,767]     # total value is calculated and the chart is presented in form of percentage%
colour=['r','g','c','b']
explode=[0.1,0,0,0] #distance of the chart outside the chart
plt.pie(size,explode=explode,colors=colour,autopct="%0.f%%",labels=label,shadow=True)
#for labels we have to give list as input
#plt.axis('equal')
plt.show()
```
