# Seaborn

 Seaborn is a visualization technique where we have a dataset and we plot a graph according to the correlation between the 
 feature values 
 
In a given data set we will have to divide the data into features particularly 
Independent Features and Dependent Feature

The plots are done with respect to the values of the features
Basically it is plotted between int and float type of data if the value is object then we use categoral features

```python
import seaborn as sns
```

```python
sns.get_dataset_names() #displays all the inbulit dataset present in seaborn
```

```python
#Correlation with heatmap

df=sns.load_dataset("iris")
df1=sns.load_dataset('tips')
```

```python
print(df.head())
df1.head()
```

```python
df.dtypes  # gives the value of the feature for comparing i.e if its a integer or float or object
```

```python
df.corr() #this ranges from (-1 to +1) it is called coefficient correlation
          #gives output of onlyinteger and floating values
```

```python
sns.heatmap(df.corr()) #heatmap is used to get color output of the data set matrix displayed above

#it is used to do feature selection when there are many features with similar values we select one vales insted of all the values
```

```python
sns.heatmap(df1.corr())
```

```python
There are 3 types of plots for integer and floating values
jointPlot, pairPlot, distPlot
```

```python
# JointPlot --- it should have only 2 values
sns.jointplot(x="sepal_length",y='petal_length',data=df,kind='hex')
```

```python
sns.jointplot(x="petal_length",y='petal_width',kind='reg',data=df)
#'reg' is regression plotting --- It draws a ling on probability density function
```

```python
#PairPlot(Scatter Plot)
sns.pairplot(df)
```

```python
sns.pairplot(df1,hue='sex') #'hue'------is the parameter to use for comparing one particular feature
```

```python
#DistPlot ----- It gives histogram output
sns.distplot(df1['tip'])
# by default the KernelDensityEstimation(KDE is True) thats the reason we get y axis in percentage values and o/p wrt count
```

```python
sns.distplot(df1['tip'],kde=False,bins=10)
#as we set KDE=False we get o/p wrt density and bins is equal dividing of the values
```

# Categorical Plots

```python
Its mainly ued to plot object values including float and integer
Types of plots: BarPlot, BoxPlot, Violin Plot, CountPlot
```

```python
#CountPlot -- it gives number of count in categorical bin(division/type) using bar
sns.countplot('sex',data=df1)
#only one axis/feature is specified in this plot we can mention the axis for it to consider
```

```python
sns.countplot(y='sex',data=df1)
```

```python
#BarPlot
sns.barplot(x="species",y='sepal_length',data=df)
```

```python
sns.barplot(x="sepal_length",y='petal_length',data=df)
```

```python
#BoxPlot
sns.boxplot(x="species",y='sepal_length',data=df)
```

```python
sns.boxplot('total_bill',"sex",hue='smoker',data=df1)
```

```python
#ViolinPlot ----- combination of both box plot inside the distribution on the top as feather like structure
sns.violinplot(x="species",y='sepal_length',data=df,palette='rainbow')
```

```python
sns.violinplot('total_bill',"day",data=df1,palette='rainbow')
#palette is the parameter used to give colour
```
